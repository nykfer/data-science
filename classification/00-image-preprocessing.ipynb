# Resize images

```python
import cv2
import os

def get_resized_image(folder_name):
    folder_path = os.path.join(os.getcwd(), folder_name)
    size = (240, 240)
    resized_folder = os.path.join(folder_path, 'resized')
    os.makedirs(resized_folder, exist_ok=True)

    for name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, name)
        # Skip the 'resized' folder itself if present
        if name == 'resized':
            continue
        if os.path.isfile(file_path):
            try:
                image = cv2.imread(file_path)
                if image is None:
                    print(f'Error: Unable to read image file {file_path}')
                    continue
                # Resize image to the target size, regardless of current dimensions
                resized = cv2.resize(image, size, interpolation=cv2.INTER_LINEAR)
                # Save resized image
                save_path = os.path.join(resized_folder, name)
                cv2.imwrite(save_path, resized)
            except Exception as e:
                print(f'Error processing {file_path}: {e}')

# Get resized images
get_resized_image('cats-dataset/cat')
get_resized_image('dogs-dataset/dog')
```
